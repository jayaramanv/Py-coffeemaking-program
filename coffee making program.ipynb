```python
class CoffeeMachine:
    def __init__(self):
        self.water = 1000
        self.milk = 500
        self.coffee = 250
        self.money = 0

    def prompt_user(self):
        self.choice = input("What would you like? (espresso/latte/cappuccino): ").lower()
        if self.choice == "off":
            exit()
        elif self.choice == "report":
            self.print_report()
            self.prompt_user()
        else:
            self.check_resources()

    def check_resources(self):
        if self.choice == "espresso":
            if self.water < 50 or self.coffee < 20:
                print("Sorry, there is not enough water or coffee.")
                self.prompt_user()
            else:
                self.process_coins(1.5)
        elif self.choice == "latte":
            if self.water < 200 or self.milk < 150 or self.coffee < 24:
                print("Sorry, there is not enough water, milk or coffee.")
                self.prompt_user()
            else:
                self.process_coins(2.5)
        elif self.choice == "cappuccino":
            if self.water < 250 or self.milk < 100 or self.coffee < 24:
                print("Sorry, there is not enough water, milk or coffee.")
                self.prompt_user()
            else:
                self.process_coins(3.0)
        else:
            print("Sorry, invalid input. Please try again.")
            self.prompt_user()

    def process_coins(self, cost):
        print("Please insert coins.")
        self.quarters = int(input("How many quarters? ")) * 0.25
        self.dimes = int(input("How many dimes? ")) * 0.10
        self.nickels = int(input("How many nickels? ")) * 0.05
        self.pennies = int(input("How many pennies? ")) * 0.01
        self.total = self.quarters + self.dimes + self.nickels + self.pennies
        if self.total < cost:
            print("Sorry, that's not enough money. Money refunded.")
            self.prompt_user()
        else:
            self.money += cost
            self.change = round(self.total - cost, 2)
            if self.change > 0:
                print(f"Here is ${self.change} in change.")
            self.make_coffee()

    def make_coffee(self):
        if self.choice == "espresso":
            self.water -= 50
            self.coffee -= 20
        elif self.choice == "latte":
            self.water -= 200
            self.milk -= 150
            self.coffee -= 24
        elif self.choice == "cappuccino":
            self.water -= 250
            self.milk -= 100
            self.coffee -= 24
        print(f"Here is your {self.choice}. Enjoy!")
        self.prompt_user()

    def print_report(self):
        print(f"Water: {self.water}ml")
        print(f"Milk: {self.milk}ml")
        print(f"Coffee: {self.coffee}g")
        print(f"Money: ${self.money}")


coffee_machine = CoffeeMachine()
coffee_machine.prompt_user()
 
```
